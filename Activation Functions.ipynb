# Activation Functions
### Explantion of what they are and why they're useful

Non-linear activation functions are functions that allow models to learn non-linear relationships between input variables and a target variable. When an input to an activation function is a tensor, the activation function is applied element-wise. Activation functions contain no weights to train or optimize. If you were to stack layers of a neural network while using linear activation functions, stacking layers have relatively no effect because the network could only learn linear transformations, like linear regression is capable of.

You can feed the outputs of each layer in a neural network through an activation function. In most machine learning libraries, the activation function will be considered part of the layer.

### Sigmoid
$S(x) = \frac{1}{1+e^{-x}}$

##### Properties:
- $\lim_{x \to \infty} S(x) = 1$
- $\lim_{x \to -\infty} S(x) = 0$
- Can think of it as a neuron that fires (output of ~1) if x meets a certain threshold, else doesn't fire (output of ~0)

##### Advantages:
- Useful when you would like output values of a layer to be bounded between 0 and 1 to replicate probabilities

##### Disadvanges:
- Vanishing gradient problem can arise when the absolute value of x is large. Derivative at these x values are close to zero.
    - Derivative values range from 0 to 1. Multiplying many values in this range will push the gradient to zero.

##### Graph:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-6, 6, 1000)
y = sigmoid(x)
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Tanh
- $\text{tanh}(x) = \frac{2}{1+e^{-2x}}-1$

##### Properties:
- $\lim_{x \to \infty} S(x) = 1$
- $\lim_{x \to -\infty} S(x) = -1$

##### Disadvanges
- Also has problem of gradient descent
    - Derivative values range from 0 to 1. Multiplying many values in this range will push the gradient to zero.

##### Graph:

```python
def tanh(x):
    return 2/(1+np.exp(-2*x))-1

x = np.linspace(-6, 6, 1000)
y = tanh(x)
plt.plot(x, y)
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### ReLU (Rectified Linear Unit)
- ReLU$(x)=\text{max}(0,x)$

##### Properties:
- $\lim_{x \to \infty} S(x) = \infty$
- $S(x) =0$ if $x <= 0$
- Neuron that can either fire or not fire. Widely varying magnitudes of firing (from 0 to infinity)

##### Advantages
- No issue with vanishing gradients (derivative is either 0 or 1)
- Multiplying 1*1*1*1*1 = 1.
- If, during training, the output of a ReLU neuron ever outputs a value of 0, the neuron "dies". It will always output a value of zero, 

##### Disadvantages
- If, during training, the output of a ReLU neuron ever outputs a value of 0, the neuron "dies". Gradient updates will no longer flow through this neuron.
    - Can sometimes be an advantage because it can act as a regularization technique

##### Graph:

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-6, 6, 1000)
y = relu(x)
plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Leaky ReLU
![Alt Text](https://wikimedia.org/api/rest_v1/media/math/render/svg/b164641d3e0255fe37afe613a2464f85d83dfe6e)

##### Properties:
- $\lim_{x \to \infty} S(x) = \infty$
- $\lim_{x \to -\infty} S(x) = -\infty$
- Instead evaluating to zero when x is less than zero, leaky relu evaluates to 0.01*x (or another small number instead of 0.01)

##### Advantages
- No issue with vanishing gradients (derivative is either -0.01 or 1)
- Neurons don't "die" during training or weight initialization

##### Graph:

```python
def leaky_relu(x):
    return np.where(x>0, x, 0.01*x)

x = np.linspace(-100, 100, 100000)
y = leaky_relu(x)
plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
